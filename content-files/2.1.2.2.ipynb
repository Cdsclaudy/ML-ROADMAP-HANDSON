# 2.1.2.2. DECISION TREES
## DEFINITION
**A decision tree is a graphical representation of a series of decisions and their possible outcomes**. It is a popular method for solving classification and regression problems in machine learning. In this essay, we will discuss the basic concepts, advantages, disadvantages, and applications of decision trees.

## COMPONENTS
A decision tree consists of three types of nodes:

- **Root node**: The topmost node that represents the entire data set or the initial problem.
- **Internal node**: A node that splits the data into two or more subsets based on a feature or a condition.
- **Leaf node**: A terminal node that represents the final outcome or the predicted value.

## SPLITTING CRITERIA
A decision tree can be constructed by using various algorithms, such as ID3, C4.5, CART, etc. These algorithms use different criteria to select the best feature or condition to split the data at each node. Some of the common criteria are:

- **Entropy: A measure of randomness or uncertainty in the data**. A low entropy means the data is pure or homogeneous, while a high entropy means the data is impure or heterogeneous.
- **Information gain: A measure of how much information a feature or a condition provides about the outcome**. It is calculated as the difference between the entropy before and after the split. A high information gain means the feature or condition reduces the uncertainty about the outcome.
- **Gini index: A measure of how well a feature or condition separates the data into different classes**. It is calculated as the sum of squared probabilities of each class in each subset. A low Gini index means the feature or condition creates pure or homogeneous subsets.

## BUILDING A DECISION TREE
The process of building a decision tree involves recursively splitting the data until one of the following conditions is met:

- All the instances in a subset belong to the same class or have the same value.
- There are no more features or conditions to split the data.
- The size of a subset is smaller than a predefined threshold.

## CLASSIFICATION PROCESS
The process of using a decision tree for prediction involves traversing the tree from the root node to a leaf node based on the values of the features or conditions at each node.

## ADVANTAGES

- They are easy to understand and interpret, as they mimic human reasoning.
- They can handle both numerical and categorical features, and can also deal with missing values.
- They are flexible and can capture non-linear relationships in the data.
- They are robust to outliers and noise in the data.
Some of the disadvantages of decision trees are:

## DISADVANTAGES

- They are prone to overfitting, especially when the tree is too deep or complex. This means they may perform well on the training data but poorly on new or unseen data.
- They are sensitive to small changes in the data, which may result in a different tree structure.
- They may create biased trees if some classes or features dominate the data set.

## APPLICATIONS

- **Medical diagnosis**: Decision trees can be used to diagnose diseases based on symptoms and test results.
- **Customer segmentation**: Decision trees can be used to segment customers based on their demographics, preferences, behavior, etc.
- **Fraud detection**: Decision trees can be used to detect fraudulent transactions based on patterns and anomalies in the data.
- **Natural language processing**: Decision trees can be used to perform tasks such as sentiment analysis, text classification, part-of-speech tagging, etc.

## CONCLUSION
In conclusion, decision trees are powerful and versatile tools for solving classification and regression problems in machine learning. They have a simple and intuitive structure that can represent complex decisions and outcomes. However, they also have some limitations and challenges that need to be addressed carefully. Decision trees are widely used in various domains and applications that require decision making and prediction.

## HANDS-ON: DECISION TREES

### 1. IMPORT LIBRARIES

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
from sklearn.model_selection import train_test_split
```

### 2. DATASET

```python
iris = load_iris()
X = iris.data
y = iris.target
```

### 3. PREPROCESSING

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

### 4. DECISION TREE

```python
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

### 5. VISUALIZING TREE

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,8))
plot_tree(clf, filled=True)
plt.show()
```
